```python
import pandas as pd
from pptx import Presentation
from pptx.util import Inches
from datetime import datetime
from PIL import Image, ImageDraw
import os
```

```python
TEMPLATE_PATH = "template.pptx"
OUTPUT_FOLDER = "output/"

IMAGE_FOLDER = "images/"
TEMP_FOLDER = "images/temp/"

# Birthday slide index (based on your PPT)
BIRTHDAY_SLIDE_INDEX = 5

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
os.makedirs(TEMP_FOLDER, exist_ok=True)
```

```python
df = pd.read_csv("employees.csv")

df['dob'] = pd.to_datetime(df['dob'], format='%d-%m-%Y')

today = datetime.today()
```

```python
birthday_df = df[
    (df['dob'].dt.day == today.day) &
    (df['dob'].dt.month == today.month)
]

print(f"🎂 Found {len(birthday_df)} birthdays today")
birthday_df
```

```python
def make_circle_image(input_path, output_path):
    img = Image.open(input_path).convert("RGBA")

    # Crop to square
    size = min(img.size)
    left = (img.width - size) // 2
    top = (img.height - size) // 2
    right = left + size
    bottom = top + size

    img = img.crop((left, top, right, bottom))

    # Create circular mask
    mask = Image.new("L", (size, size), 0)
    draw = ImageDraw.Draw(mask)
    draw.ellipse((0, 0, size, size), fill=255)

    img.putalpha(mask)
    img.save(output_path)
```

```python
def prepare_employee_image(emp_code):
    input_path = f"{IMAGE_FOLDER}{emp_code}.jpg"
    output_path = f"{TEMP_FOLDER}{emp_code}.png"

    if os.path.exists(input_path):
        make_circle_image(input_path, output_path)
        return output_path
    else:
        print(f"⚠️ Missing image for {emp_code}")
        return None
```

```python
def replace_name(slide, name):
    for shape in slide.shapes:
        if shape.has_text_frame:
            for paragraph in shape.text_frame.paragraphs:
                for run in paragraph.runs:
                    if "{NAME}" in run.text:
                        run.text = run.text.replace("{NAME}", name)
```

```python
from pptx.util import Cm

def add_circular_image(slide, img_path):
    if img_path:
        slide.shapes.add_picture(
            img_path,
            Cm(3.28),   # left
            Cm(6.92),   # top
            width=Cm(8.4),
            height=Cm(8.4)
        )
```

```python
for _, row in birthday_df.iterrows():

    print(f"🎉 Processing birthday for {row['name']}")

    # Load fresh template (IMPORTANT)
    prs = Presentation(TEMPLATE_PATH)
    slide = prs.slides[BIRTHDAY_SLIDE_INDEX]

    # Replace name
    replace_name(slide, row['name'])

    # Prepare circular image
    img_path = prepare_employee_image(row['employee_code'])

    # Add image
    add_circular_image(slide, img_path)

    # Remove other slides (keep only birthday slide)
    slides_to_remove = []
    for i in range(len(prs.slides)):
        if i != BIRTHDAY_SLIDE_INDEX:
            slides_to_remove.append(i)

    for i in reversed(slides_to_remove):
        rId = prs.slides._sldIdLst[i].rId
        prs.part.drop_rel(rId)
        del prs.slides._sldIdLst[i]

    # Save file
    output_file = f"{OUTPUT_FOLDER}birthday_{row['employee_code']}.pptx"
    prs.save(output_file)

    print(f"✅ Saved: {output_file}")
```

```python
from pptx.util import Cm

def add_employee_image(slide, img_path):
    if img_path:
        slide.shapes.add_picture(
            img_path,
            Cm(16.33),   # left
            Cm(4.67),    # top
            width=Cm(8.5),
            height=Cm(9.45)
        )
```

```python
from pptx.util import Cm

def add_year_image(slide, years):

    # ❌ Skip 5 & 10
    if years in [5, 10]:
        return

    digits = list(str(years))

    width = 1.24
    height = 2.15

    # 🔹 SINGLE DIGIT
    if len(digits) == 1:

        digit = digits[0]
        path = f"images/years/Picture{digit}.png"

        if not os.path.exists(path):
            print(f"⚠️ Missing: Picture{digit}.png")
            return

        slide.shapes.add_picture(
            path,
            Cm(6.39),   # single digit position
            Cm(2.75),
            width=Cm(width),
            height=Cm(height)
        )

    # 🔹 DOUBLE DIGIT
    elif len(digits) == 2:

        positions = [
            (5.9, 2.6),   # first digit
            (6.88, 2.58)  # second digit
        ]

        for i, digit in enumerate(digits):

            path = f"images/years/{digit}.png"

            if not os.path.exists(path):
                print(f"⚠️ Missing: {digit}.png")
                continue

            left, top = positions[i]

            slide.shapes.add_picture(
                path,
                Cm(left),
                Cm(top),
                width=Cm(width),
                height=Cm(height)
            )
```

```python
def is_anniversary(row):
    return (
        row['doj'].day == today.day and
        row['doj'].month == today.month
    )

def get_years(row):
    return today.year - row['doj'].year
```

```python
df['doj'] = pd.to_datetime(df['doj'], format='%d-%m-%Y', errors='coerce')
df['dob'] = pd.to_datetime(df['dob'], format='%d-%m-%Y', errors='coerce')
for _, row in df.iterrows():

    if not is_anniversary(row):
        continue

    # Only junior for now
    if row['grade'] not in ["JM", "Analyst"]:
        continue

    years = get_years(row)

    print(f"🎉 Processing: {row['name']} ({years} years)")

    # Load template
    prs = Presentation(TEMPLATE_PATH)

    # 🎯 STEP 1: Choose slide
    if years == 5:
        slide_index = 2
    elif years == 10:
        slide_index = 3
    else:
        slide_index = 0   # junior normal

    slide = prs.slides[slide_index]

    # 🎯 STEP 2: Replace name
    replace_name(slide, row['name'])

    # 🎯 STEP 3: Add employee image (different for milestone vs normal)

    img_path = f"images/{row['employee_code']}.jpg"


    if years == 5:
        slide.shapes.add_picture(
            img_path,
            Cm(17.41),
            Cm(3.84),
            width=Cm(7.32),
            height=Cm(8.03)
    )

    elif years == 10:
        slide.shapes.add_picture(
            img_path,
            Cm(16.13),
            Cm(4.71),
            width=Cm(7.37),
            height=Cm(7.37)
        )    
    else:
        # Normal junior image
        slide.shapes.add_picture(
            img_path,
            Cm(16.33),
            Cm(4.67),
            width=Cm(8.5),
            height=Cm(9.45)
        )

    # 🎯 STEP 4: Add year digits ONLY if not milestone
    if years not in [5, 10]:
        add_year_image(slide, years)

    # 🎯 STEP 5: Remove other slides
    slides_to_remove = []
    for i in range(len(prs.slides)):
        if i != slide_index:
            slides_to_remove.append(i)

    for i in reversed(slides_to_remove):
        rId = prs.slides._sldIdLst[i].rId
        prs.part.drop_rel(rId)
        del prs.slides._sldIdLst[i]

    # 🎯 STEP 6: Save
    output_file = f"{OUTPUT_FOLDER}anniv_{row['employee_code']}.pptx"
    prs.save(output_file)

    print(f"✅ Saved: {output_file}")
    
```

```python
from pptx.util import Cm

def add_senior_image(slide, img_path):
    if img_path:
        pic = slide.shapes.add_picture(
            img_path,
            Cm(9.27),
            Cm(3.87),
            width=Cm(11.87),
            height=Cm(11.76)
        )
        return pic
```

```python
def send_to_back(shape):
    sp = shape._element
    parent = sp.getparent()
    parent.remove(sp)
    parent.insert(0, sp)   
```

```python
def add_year_image_senior(slide, years):

    # Skip milestones
    if years in [5, 10]:
        return

    digits = list(str(years))

    width = 1.24
    height = 2.15

    # 🔹 SINGLE DIGIT
    if len(digits) == 1:

        digit = digits[0]
        path = f"images/years/Picture{digit}.png"

        if not os.path.exists(path):
            print(f"⚠️ Missing: {digit}.png")
            return

        slide.shapes.add_picture(
            path,
            Cm(6.24),
            Cm(2.58),
            width=Cm(width),
            height=Cm(height)
        )

    # 🔹 DOUBLE DIGIT
    elif len(digits) == 2:

        positions = [
            (5.6, 2.53),   # first digit
            (6.58, 2.53)   # second digit (slightly spaced)
        ]

        for i, digit in enumerate(digits):

            path = f"images/years/Picture{digit}.png"

            if not os.path.exists(path):
                print(f"⚠️ Missing: Picture{digit}.png")
                continue

            left, top = positions[i]

            slide.shapes.add_picture(
                path,
                Cm(left),
                Cm(top),
                width=Cm(width),
                height=Cm(height)
            )
```

```python
for _, row in df.iterrows():

    if not is_anniversary(row):
        continue

    # 🎯 Only Senior
    if row['grade'] not in ["EVP", "Partner"]:
        continue

    years = get_years(row)

    print(f"👑 Senior Anniversary: {row['name']} ({years} years)")

    prs = Presentation(TEMPLATE_PATH)

    # Slide selection
    if years == 5:
        slide_index = 2
    elif years == 10:
        slide_index = 3
    else:
        slide_index = 1   # senior slide

    slide = prs.slides[slide_index]

    # Replace name
    replace_name(slide, row['name'])

    # Circular image (reuse birthday logic)
    img_path = prepare_employee_image(row['employee_code'])

    # Add image
    pic = add_senior_image(slide, img_path)

    # 🔥 Send behind overlay
    send_to_back(pic)
    # Add numbers ONLY if not milestone
    if years not in [5, 10]:
        add_year_image_senior(slide, years)
    # Remove other slides
    slides_to_remove = []
    for i in range(len(prs.slides)):
        if i != slide_index:
            slides_to_remove.append(i)

    for i in reversed(slides_to_remove):
        rId = prs.slides._sldIdLst[i].rId
        prs.part.drop_rel(rId)
        del prs.slides._sldIdLst[i]

    # Save
    output_file = f"{OUTPUT_FOLDER}anniv_senior_{row['employee_code']}.pptx"
    prs.save(output_file)

    print(f"✅ Saved: {output_file}")
```
